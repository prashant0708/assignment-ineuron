# Q-1. Write an SQL query to fetch “FIRST_NAME” from Worker table using the alias name as WORKER_NAME

```python
import pandas as pd
pd.read_sql('select FIRST_NAME as WORKER_NAME  from employees' ,mydb)
```

# Q-3. Write an SQL query to show the last 5 record from a table.

```python
##Showing all the records 
pd.read_sql('select * from employees  ' ,mydb)
```

```python
#show the last 5 record from a table.
pd.read_sql('select * from (select * from employees order by emp_no desc limit 5) as a order by emp_no asc  ',mydb)
```
